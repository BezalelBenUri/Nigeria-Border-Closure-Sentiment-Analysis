```python
import numpy as np
import pandas as pd
import re

from transformers import pipeline

tweet_df = pd.read_csv('twitterOpsBC.csv')
tweet_df.head()
```

```python
def cleanText(text):
    text = re.sub(r'@[A-Za-z0-9]+', '', text)
    text = re.sub(r'#', '', text)
    text = re.sub(r'RT[\s]+', '', text)
    text = re.sub(r'https?:\/\/\S+', '', text)
    return text

tweet_df['tweet'] = tweet_df['tweet'].apply(cleanText)
tweet_df.head()
```

```python
sentiment_classifier = pipeline('sentiment-analysis')
result = sentiment_classifier(tweet_df['tweet'].tolist())


sentiment =[]
score = []
ind = 0
for result in result:
    print('Analyzing...')
    print(tweet_df['tweet'][ind] + '/n')
    print(f"label: {result['label']} with score : {round(result['score'], 4)}")
    sentiment.append(result['label'])
    score.append(round(result['score'], 4))
    print('===\n')
    
    ind = ind + 1
```

```python
tweet_df['Sentiment'] = sentiment
tweet_df['Score'] = score
tweet_df.head()
```

```python
tweet_df.to_csv('twitterSentiment.csv', encoding = 'utf-8')
```
