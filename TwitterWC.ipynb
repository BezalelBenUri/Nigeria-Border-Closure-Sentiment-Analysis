```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from wordcloud import WordCloud, STOPWORDS

data = pd.read_csv('twitterOpsBC.csv')



plt.figure(figsize = (40, 20)) 

text = " ".join(review for review in data.tweet.astype(str))
print(len(text))
```

```python
stopwords = set(STOPWORDS)
stopwords.update(['', 'nigeria', 'nigerian', 'border', 'borders','closure', 'closures' 'will', 'still','said', 'country', 'https', 's', 'now', 'amp', 't', 'co', 'u', 
                  'every', 'closed', 'one'])
wordcloud1 = WordCloud(stopwords = stopwords).generate(text)

    
plt.figure(figsize = (20, 9))
plt.imshow(wordcloud1)
plt.axis("off")
plt.show()
```
