```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from wordcloud import WordCloud, STOPWORDS

data = pd.read_csv('NairalandBorderClosure.csv')



plt.figure(figsize = (12,12)) 

text = " ".join(review for review in data.Comment.astype(str))
print(len(text))
```

```python
stopwords = set(STOPWORDS)
stopwords.update(['â', 'nigeria', 'nigerian', 'border', 'closure', 'will', 'said', 'country'])
wordcloud1 = WordCloud(stopwords = stopwords).generate(text)

    
plt.figure(figsize = (20, 9))
plt.imshow(wordcloud1)
plt.axis("off")
plt.show()
```
